```python
def clean_text(text): 
    pat1 = r'@[^ ]+'                   
    pat2 = r'https?://[A-Za-z0-9./]+'  
    pat3 = r'\'s'                      
    pat4 = r'\#\w+'                     
    pat5 = r'&amp '                     
    pat6 = r'[^A-Za-z\s]'               
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    text = [word for word in text.split() if word.lower() not in exclude]
    text = " ".join(text)
    return text.strip()
```

```python
data_df["cleaned_tweet"] = data_df["tweet"].apply(clean_text)
```

```python
data_df["cleaned_tweet"].head()
```

```python
for row in data_df.itertuples():
    tweet = data_df.at[row[0], 'cleaned_tweet']
#run sentiment using TextBlob
    analysis = TextBlob(tweet)
#set value to dataframe
    data_df.at[row[0], 'polarity'] = analysis.sentiment[0]
    data_df.at[row[0], 'subjectivity'] = analysis.sentiment[1]
#Create Positive / negative column depending on polarity
    if analysis.sentiment[0]>0:
        data_df.at[row[0], 'Sentiment'] = "Positive"
    elif analysis.sentiment[0]<0:
        data_df.at[row[0], 'Sentiment'] = "Negative"
    else:
        data_df.at[row[0], 'Sentiment'] = "Neutral"
```

```python
from sklearn import metrics
from sklearn import preprocessing
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation

feature_cols = ['changeInValueUSD', 'previousValueUSD', 'ratio', 'polarity']

X = data[feature_cols]
y = data.value

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)


label_encoder = preprocessing.LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.fit_transform(y_test)

clf=RandomForestClassifier(n_estimators=100)
clf = clf.fit(X_train, y_train_enc)
    
y_pred = clf.predict(X_test)

print("R-squared score:", metrics.r2_score(y_test_enc, y_pred))    
```
